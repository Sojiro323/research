```python
def get_precision(precision):
    count = [0,0,0]
    for i in range(1,40):
        count[0] += precision[str(i)][1]
        count[1] += precision[str(i)][2]
        count[2] += precision[str(i)][3]

    return ((count[0] * 1.0) + (count[1] * 0.5)) / (count[0] + count[1] + count[2])
```

```python
DIR = '../query/'
seeds_score = Mypickle.load(DIR, 'seeds_score_fx_australia_0918_sum')
len(seeds_score)
```

```python
queryID = "0909_mutual"
now_seeds = database.select("select userID from {0} where queryID = \'{1}\' and (result=\'2\' or result=\'None\')".format("prof", queryID))
len(now_seeds)
```

```python
ranking_list = []
start_score = [0.25, 1, 0, 3]
for seed_k, seed_v in seeds_score.items():
    precision = get_precision(seed_v["precision"])
    for path_k, path_v in seed_v["score"].items():
        if seed_v["precision"][path_k][1] == start_score[1] and seed_v["precision"][path_k][2] == start_score[2] and seed_v["precision"][path_k][3] == start_score[3]:
            ranking_list.append([seed_k,path_k,(precision*seed_v["molecule"][path_k][0])/seed_v["denominator"][path_k][0]])

        else:
            ranking_list.append([seed_k,path_k,seed_v["precision"][path_k][0]])
ranking_list.sort(key=lambda x:x[2])
ranking_list.reverse()
for i,v in enumerate(ranking_list):
    print("{0}. {1}  {2}  {3}".format(i+1, v[2], v[0],v[1]))
```

```python
for sk,sv in seeds_score.items():
    molecule = sv["molecule"]
    denominator = sv["denominator"]
    precision = sv['precision']
    score = sv['score']
    find = sv['find']
    print("seed :", sk)
    for i in range(1,40):
        print(i,molecule[str(i)], denominator[str(i)],precision[str(i)],score[str(i)],find[str(i)])
    print('\n')
```
